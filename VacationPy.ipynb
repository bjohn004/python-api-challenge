```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import gmaps
import os
import time

# Import API key
from api_keys import g_key
```

```python
# import csv from part 1 of homework
csv_w = "cities_submit.csv"
csv_weather = pd.read_csv(csv_w)
csv_weather
```

```python
# configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# drop na values from weather csv and find max humidity for use in heat map
csv_weather = csv_weather.dropna()
csv_weather["Humidity"].max()
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = csv_weather[["Latitude", "Longitude"]].astype(float)
```

```python
# find humidity column, make series
humidity = csv_weather["Humidity"].astype(float)
```

```python
 # Create a humidity Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 5)

fig.add_layer(heat_layer)

fig
```

```python
# create a DF that has ideal weather conditions
df_ideal_temp = csv_weather.loc[(csv_weather["Temp"] >= 65) & (csv_weather["Temp"] <=75), :] 
df_ideal_wind= df_ideal_temp.loc[(df_ideal_temp["Wind Speed"] >= 5) & (df_ideal_temp["Wind Speed"] <=15), :] 
df_ideal = df_ideal_wind.loc[(df_ideal_wind["Cloudiness"] >= 0) & (df_ideal_wind["Cloudiness"] <=50), :] 
df_ideal
```

```python
# create new simplified DF to add hotel name to
hotel_df = df_ideal[["City", "Country", "Latitude", "Longitude"]]
hotel_df["Hotel Name"] = ""
hotel_df = hotel_df.reset_index()
del hotel_df["index"]
hotel_df
```

```python
# create empty list to be filled in following for loop 
hotel_name = []
```

```python
# define base url and parameters for searching within the gmap api
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "radius": 5000,
    "keyword": "hotel",
    "types": "hotel",
    "key": g_key
}

# complete loop through the DF to find closest hotel to the provided city center coordinates
for index, row in hotel_df.iterrows():
    city = row["City"]
    print(f"Retrieving Results for {index}, the closest hotel to {city} is: ")
    lat = row["Latitude"]
    lng = row["Longitude"]
    params["location"] = f"{lat}, {lng}"
    response = requests.get(base_url, params = params)
    hotel_data = response.json()
    
    # try and except the results from the above API request and append the list with successful requests
    try:
        hotel_name.append(hotel_data["results"][0]["name"])
        print(f"{hotel_name[index]}")
    except:
        print("------------")
        print("No Hotel found") 
        print("------------")
        hotel_name.append("NO HOTEL NEARBY")
```

```python
# add hotel list to the empty column in the dataframe
hotel_df["Hotel Name"] = hotel_name
hotel_df
```

```python
 # redisplay humidity Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 5)

fig.add_layer(heat_layer)

fig
```

```python
# create DF for LAT/LONG to be used in graph
locations_hotels = hotel_df[["Latitude", "Longitude"]].astype(float)
locations_hotels
```

```python
# use info box template provided in starter code
info_box_template= """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# create symbol layer for locations of hotels that meet ideal conditions
hotel_layer = gmaps.symbol_layer(
    locations_hotels, fill_color='rgba(150, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=7,
    info_box_content= hotel_info)
fig = gmaps.figure()
fig.add_layer(hotel_layer)

fig
```

```python
 # Create a combined map
fig = gmaps.figure()

fig.add_layer(heat_layer)
fig.add_layer(hotel_layer)

fig
```
